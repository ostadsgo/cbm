```python
import pandas as pd


pd.set_option('display.max_columns', None)

# Read csv
df = pd.read_csv("LABORATORY_DATA.csv")
df_date = pd.read_csv("datetime.csv")

# merge month and year with the data
df = pd.merge(df, df_date[['PointID', 'Probenahme_Monat', 'Probenahme_Jahr']], 
                    on='PointID', how='left')

# Move `Month` and `Year` column 
month = df.pop("Probenahme_Monat")
year = df.pop("Probenahme_Jahr")
df.insert(1, "Month", month)
df.insert(2, "Year", year)

# Save the df with month and year to new csv file
df.to_excel("lab_data_date.xlsx", index=False)

print(len(df))

# df.head()


```

# Data Cleaning

```python
# Remove null values based on `Water content` and `TOC` columns.
df = df.dropna(subset=["Water content", "TOC"])

# Water content > 0 and < 100 does not make sense
df = df[(df["Water content"] >= 0) & (df["Water content"] <= 100)]

# Water content > 0 and < 500 does not make sense
df = df[(df["TOC"] >= 0) & (df["TOC"] <= 500)]


print(len(df))


# 17189 - x  = 16610
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your existing code
plt.figure(figsize=(10, 6))
plt.scatter(df['TOC'], df['Water content'], color='blue')

# Calculate linear regression
slope, intercept = np.polyfit(df['TOC'], df['Water content'], 1)

# Create regression line
regression_line = slope * df['TOC'] + intercept

# Plot regression line
plt.plot(df['TOC'], regression_line, color='red', label=f'Linear Regression: y = {slope:.2f}x + {intercept:.2f}')

plt.title('Scatter Plot of TOC vs Water Content with Linear Regression')
plt.xlabel('Total Organic Carbon (TOC)')
plt.ylabel('Water Content')
plt.grid()
plt.legend()
plt.show()
```

```python
# correlation
# Step 3: Calculate Correlation
correlation = df['TOC'].corr(df['Water content'])
print(f'Correlation between TOC and Water Content: {correlation:.2f}')
```

```python
import statsmodels.api as sm

df_1 = df[["TOC", "Water content", "BD_bulk", "TN", "pH_H2O" ]]
df_2 = df_1.dropna()
# Define the independent variable (X) and dependent variable (y)
X = df_2[['Water content', "BD_bulk"]]  # Independent variable
y = df_2['TOC']            # Dependent variable

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())
```

# Key Components of the OLS Regression Results

    Dependent Variable:
        Dep. Variable: TOC: This indicates that the model is predicting Total Organic Carbon (TOC).

    Model Fit Statistics:
        R-squared: 0.443: This value indicates that approximately 44.3% of the variability in TOC can be explained by the water content. While this is a moderate level of explanation, it suggests that other factors may also influence TOC.
        Adj. R-squared: 0.437: This adjusted value accounts for the number of predictors in the model. It is slightly lower than R-squared, which is common when using a single predictor.

    F-statistic: 83.41: This statistic tests the overall significance of the model. A high F-statistic indicates that the model is statistically significant.
        Prob (F-statistic): 5.39e-15: This p-value is extremely small (close to 0), indicating that the model is statistically significant, and at least one predictor (in this case, water content) is significantly related to the dependent variable (TOC).

    Coefficients:
        const (Intercept): -5.4044: This is the estimated value of TOC when water content is zero. In practical terms, it may not have a meaningful interpretation if zero water content is not realistic in your context.
        Water content: 0.7040: This coefficient indicates that for each unit increase in water content, TOC is expected to increase by approximately 0.704 units. This suggests a positive relationship between water content and TOC.

    Standard Errors:
        std err: These values indicate the standard error of the coefficients. A smaller standard error relative to the coefficient suggests a more precise estimate.
        For example, the standard error for the water content coefficient is 0.077.

    t-statistic and P-value:
        t: The t-statistic for water content is 9.133, which is quite high, indicating that the coefficient is significantly different from zero.
        P>|t|: The p-value for water content is 0.000, which is less than 0.05, indicating that the relationship between water content and TOC is statistically significant.

    Confidence Intervals:
        [0.025, 0.975]: This range provides the 95% confidence interval for the coefficients. For water content, the interval is [0.551, 0.857], meaning we are 95% confident that the true coefficient lies within this range.

    Residual Statistics:
        Omnibus: A test for the normality of residuals. A high p-value (0.854) suggests that the residuals are normally distributed.
        Durbin-Watson: 1.227: This statistic tests for autocorrelation in the residuals. Values close to 2 suggest no autocorrelation. Values below 2 may indicate positive autocorrelation.
        Skew: -0.111: This value indicates the symmetry of the residuals. A value close to 0 suggests a normal distribution.
        Kurtosis: 2.978: This value indicates the "tailedness" of the distribution. A value close to 3 suggests a normal distribution.

Summary

    The regression analysis indicates a statistically significant positive relationship between water content and TOC, with water content explaining about 44.3% of the variability in TOC.
    The model is statistically significant, and the coefficient for water content suggests that increases in water content are associated with increases in TOC.
    While the model provides useful insights, the R-squared value indicates that there are likely other factors influencing TOC that are not included in this model. Further analysis may be needed to explore these additional factors.

```python
from scipy import stats

# Calculate correlation between TOC and Water Content for each soil texture class
results = []
for texture_class in df['Soil texture class'].unique():
    subset = df[df['Soil texture class'] == texture_class]
    if len(subset) > 1:  # Ensure enough data points for correlation
        corr, p_value = stats.pearsonr(subset['TOC'], subset['Water content'])
        results.append({
            'Soil texture class': texture_class,
            'Correlation': corr,
            'P_Value': p_value,
            'Sample_Size': len(subset)
        })

# Create results DataFrame
correlation_results = pd.DataFrame(results)
print(correlation_results)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Define your layer ranges
layer_ranges = [
    (0, 10),
    (10, 30),
    (30, 50),
    (50, 70),
    (70, 100)
]

# Create a list of your DataFrames
dfs = []
for upper, lower in layer_ranges:
    df_layer = df_sand[(df_sand["Layer upper limit"] == upper) & (df_sand["Layer lower limit"] <= lower)]
    dfs.append((f"{upper}_{lower}", df_layer))


# Find global min and max values across all DataFrames
x_min, x_max = float('inf'), float('-inf')
y_min, y_max = float('inf'), float('-inf')

for name, dfx in dfs:
    if len(dfx) > 0:
        x_min = min(x_min, dfx['TOC'].min())
        x_max = max(x_max, dfx['TOC'].max())
        y_min = min(y_min, dfx['Water content'].min())
        y_max = max(y_max, dfx['Water content'].max())

# Add some padding to the axis limits
x_padding = (x_max - x_min) * 0.1 if x_max > x_min else 0.1
y_padding = (y_max - y_min) * 0.1 if y_max > y_min else 0.1

x_min -= x_padding
x_max += x_padding
y_min -= y_padding
y_max += y_padding


# Create subplots
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()  # Flatten the 2D array of axes for easier indexing

# Colors for each plot
colors = ['blue', 'green', 'red', 'purple', 'orange']

# Plot each DataFrame
for i, (name, dfx) in enumerate(dfs):
    if i >= len(axes):  # Ensure we don't exceed the number of subplots
        break
        
    ax = axes[i]
    
    if len(dfx) > 0:  # Only plot if DataFrame is not empty
        # Scatter plot
        ax.scatter(dfx['TOC'], dfx['Water content'], color=colors[i], alpha=0.7)
        
        # Calculate and plot linear regression if we have enough data points
        if len(dfx) > 1:
            slope, intercept = np.polyfit(dfx['TOC'], dfx['Water content'], 1)
            regression_line = slope * dfx['TOC'] + intercept
            ax.plot(dfx['TOC'], regression_line, color='red', 
                   label=f'y = {slope:.2f}x + {intercept:.2f}')
            
        ax.set_title(f'Layer {name.replace("_", "-")} cm')
        ax.set_xlabel('Total Organic Carbon (TOC)')
        ax.set_ylabel('Water Content')
        ax.grid(True, alpha=0.3)
        ax.legend()
        
        # Set the same axis limits for all plots
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
    else:
        ax.text(0.5, 0.5, 'No data available', ha='center', va='center', transform=ax.transAxes)
        ax.set_title(f'Layer {name.replace("_", "-")} cm')
        
        # Set the same axis limits for all plots
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)

# Remove any empty subplots if we have fewer than 6 DataFrames
if len(dfs) < len(axes):
    for j in range(len(dfs), len(axes)):
        fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
