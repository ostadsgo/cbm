```python
import pandas as pd
from scipy.stats import pearsonr


def calculate_r(x, y, soil_type="", layer=""):

    # Calculate correlation
    correlation, p_value = pearsonr(x, y)
    
    # Determine strength and direction
    if abs(correlation) >= 0.7:
        strength = "strong"
    elif abs(correlation) >= 0.3:
        strength = "moderate"
    else:
        strength = "weak"
    
    direction = "positive" if correlation > 0 else "negative"
    significant = p_value < 0.05

    # Return as dictionary for DataFrame creation
    return {
        'soil_type': soil_type,
        'layer': layer,
        'correlation': float(f"{correlation:.4f}"),
        'p_value': float(f"{p_value:.4f}"),
        'strength': strength,
        'direction': direction,
        'significant': significant
    }


def get_soil(dfx, texture):
    """ Get each soils type which are sand, silt, clay, loam."""
    df_res = dfx[dfx['Soil texture class'].str.strip().isin(texture)]
    return df_res


def get_soils(dfx, textures):
    all_soils = {}
    for name, texture in textures.items():
        all_soils[name] = get_soil(dfx, texture)
    return all_soils

    
def get_soil_layers(df_soil, soil_name, layer_ranges):
    soil_layers = {}
    for upper, lower in layer_ranges:
        layer_name = f"{soil_name}_{upper}_{lower}"
        df_layer = df_soil[(df_soil["Layer upper limit"] >= upper) & (df_soil["Layer lower limit"] <= lower)]
        soil_layers[layer_name] = df_layer
    return soil_layers


def get_layers(dfx, layer_ranges):
    layers = {}
    for upper, lower in layer_ranges:
        layer_name = f"{upper}_{lower}"
        df_layer = dfx[(dfx["Layer upper limit"] >= upper) & (dfx["Layer lower limit"] <= lower)]
        layers[layer_name] = df_layer
    return layers



# Soil texture
sand_texture = ["Ss", "Su2", "Sl2", "Sl3", "St2", "Su3", "Su4", "Slu", "Sl4", "St3", "fS", "fSms", "mS", "mSgs", "gS", "mSfs"]
clay_texture = ["Ts4", "Ts3", "Tu3", "Tu4", "Ts2", "Tl", "Tu2", "Tt", "Ts3"]
silt_texture = ["Uu", "Us", "Ut2", "Ut3", "Uls", "Ut4"]
loam_texture = ["Ls2", "Ls3",  "Ls4",  "Lt2",  "Lts", "Lu", "Lt3"]
soil_textures = {"sand": sand_texture, "silt": silt_texture,  "clay": clay_texture,  "loam": loam_texture}

# x = 0
# for v in soil_textures.values():
#     print(len(set(v)) == len(v)) 

layer_ranges = [(0, 10), (10, 30), (30, 50), (50, 70),(70, 100)]
layer_group = [(0, 30), (30, 100)]

# df_sand = get_soil(df, sand_texture)
#get_soil_layers(df_sand, "sand", layer_group)
```

```python
from scipy.stats import pearsonr
import pandas as pd

# Calculate depth categories
def calculate_depth_category(upper, lower):
    mid_depth = (upper + lower) / 2
    if mid_depth <= 30:
        return '0-30'
    elif mid_depth <= 100:
        return '30-100'
    else:
        return '100+'


df_city = df[df['County'].str.strip().isin(['BB', 'BY', 'NI'])]  # BB, BY, NI



df_city['Depth_Category'] = df_city.apply(
    lambda row: calculate_depth_category(row['Layer upper limit'], row['Layer lower limit']), 
    axis=1
)

# Define soil categories
df_city['TOC_percent'] = df_city['TOC'] / 10
df_city['Soil_Category'] = df_city['TOC_percent'].apply(
    lambda x: 'Organic_Soil' if x > 12 else 'Mineral_Soil'
)

# Analyze correlations by depth and soil type
results = []

for depth_category in ['0-30', '30-100', '100+']:
    depth_data = df_city[df_city['Depth_Category'] == depth_category]
    
    for soil_type in ['Mineral_Soil', 'Organic_Soil']:
        soil_data = depth_data[depth_data['Soil_Category'] == soil_type]
        
        if len(soil_data) > 2:  # Need enough samples for correlation
            try:
                correlation, p_value = pearsonr(soil_data['Water content'], soil_data['TOC'])
                
                results.append({
                    'Layer': depth_category,
                    'Soil_Type': soil_type,
                    'Correlation': round(correlation, 4),
                    'P_Value': round(p_value, 4),
                    'N_Samples': len(soil_data),
                    'Mean_TOC_percent': round(soil_data['TOC_percent'].mean(), 1),
                    'Mean_Water_Content': round(soil_data['Water content'].mean(), 1),
                    'Significant': p_value < 0.05
                })
            except:
                # Handle cases where correlation can't be calculated
                results.append({
                    'Layer': depth_category,
                    'Soil_Type': soil_type,
                    'Correlation': None,
                    'P_Value': None,
                    'N_Samples': len(soil_data),
                    'Mean_TOC_percent': round(soil_data['TOC_percent'].mean(), 1),
                    'Mean_Water_Content': round(soil_data['Water content'].mean(), 1),
                    'Significant': False
                })

# Create results dataframe
results_df = pd.DataFrame(results)
#print("=== CORRECTED ANALYSIS BY DEPTH AND SOIL TYPE ===")
results_df
```

# Key Components of the OLS Regression Results

    Standard Errors:

        t: The t-statistic for water content is 9.133, which is quite high, indicating that the coefficient is significantly different from zero.

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

def perform_regression_analysis(df):
    """
    Perform comprehensive regression analysis between Water content and TOC
    """
    
    print("="*80)
    print("REGRESSION ANALYSIS: TOC as a function of Water Content")
    print("="*80)
    
    # 1. Simple Linear Regression: TOC ~ Water Content
    print("\n1. SIMPLE LINEAR REGRESSION: TOC ~ Water Content")
    print("-" * 50)
    
    # Prepare data for statsmodels
    X = df['Water content']
    y = df['TOC']
    
    # Add constant for intercept
    X = sm.add_constant(X)
    
    # Fit OLS model
    model_simple = sm.OLS(y, X).fit()
    
    print(model_simple.summary())
    
    # Extract key statistics
    r_squared_simple = model_simple.rsquared
    slope = model_simple.params['Water content']
    slope_p_value = model_simple.pvalues['Water content']
    
    print(f"\nKey Findings - Simple Regression:")
    print(f"R = {r_squared_simple:.4f} ({r_squared_simple*100:.1f}% of variance explained)")
    print(f"Slope () = {slope:.4f} (TOC change per 1% increase in water content)")
    print(f"P-value for slope: {slope_p_value:.4f}")
    
    if slope_p_value < 0.05:
        print("The relationship is statistically significant (p < 0.05)")
    else:
        print("The relationship is not statistically significant")
    
    # 2. Multiple Regression with Depth and Texture
    print("\n\n2. MULTIPLE REGRESSION: TOC ~ Water Content + Depth + Texture")
    print("-" * 60)
    
    # Check if we have the required columns
    required_columns = ['Water content', 'TOC', 'Layer upper limit', 'Layer lower limit', 'Soil type']
    missing_columns = [col for col in required_columns if col not in df.columns]
    
    if missing_columns:
        print(f"Missing columns for full analysis: {missing_columns}")
        print("Proceeding with available data...")
        return model_simple
    
    # Create depth variable (average of layer limits)
    df['Depth'] = (df['Layer upper limit'] + df['Layer lower limit']) / 2
    
    # Fit multiple regression model
    model_multiple = ols('TOC ~ Q("Water content") + Depth + C(Q("Soil type"))', data=df).fit()
    
    print(model_multiple.summary())
    
    # 3. Comparative Analysis
    print("\n\n3. COMPARATIVE MODEL ANALYSIS")
    print("-" * 40)
    
    print(f"Simple Model R: {r_squared_simple:.4f}")
    print(f"Multiple Model R: {model_multiple.rsquared:.4f}")
    
    improvement = model_multiple.rsquared - r_squared_simple
    print(f"R Improvement with additional predictors: {improvement:.4f}")
    
    # 4. Interpretation
    print("\n\n4. INTERPRETATION")
    print("-" * 30)
    
    print("The slope coefficient () represents the average change in TOC percentage")
    print("associated with a one percentage point increase in water content.")
    
    if r_squared_simple < 0.2:
        strength = "weak"
    elif r_squared_simple < 0.5:
        strength = "moderate"
    elif r_squared_simple < 0.7:
        strength = "strong"
    else:
        strength = "very strong"
    
    print(f"\nThe simple model shows a {strength} predictive relationship (R = {r_squared_simple:.3f}),")
    print(f"meaning water content explains {r_squared_simple*100:.1f}% of the variance in TOC.")
    
    if improvement > 0.1:
        print(f"\nIncluding depth and soil texture substantially improves the model")
        print(f"(R increase of {improvement:.3f}), suggesting these are important factors.")
    
    return model_simple, model_multiple

def plot_regression_results(df, model_simple):
    """
    Create visualization of regression results
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Scatter plot with regression line
    axes[0,0].scatter(df['Water content'], df['TOC'], alpha=0.6)
    
    # Generate regression line
    x_range = np.linspace(df['Water content'].min(), df['Water content'].max(), 100)
    X_pred = sm.add_constant(x_range)
    y_pred = model_simple.predict(X_pred)
    
    axes[0,0].plot(x_range, y_pred, color='red', linewidth=2)
    axes[0,0].set_xlabel('Water Content (%)')
    axes[0,0].set_ylabel('TOC (%)')
    axes[0,0].set_title(f'TOC vs Water Content\nR = {model_simple.rsquared:.3f}')
    axes[0,0].grid(True, alpha=0.3)
    
    # Residual plot
    residuals = model_simple.resid
    axes[0,1].scatter(model_simple.fittedvalues, residuals, alpha=0.6)
    axes[0,1].axhline(y=0, color='red', linestyle='--')
    axes[0,1].set_xlabel('Fitted Values')
    axes[0,1].set_ylabel('Residuals')
    axes[0,1].set_title('Residual Plot')
    axes[0,1].grid(True, alpha=0.3)
    
    # Distribution of TOC
    axes[1,0].hist(df['TOC'], bins=20, alpha=0.7, edgecolor='black')
    axes[1,0].set_xlabel('TOC (%)')
    axes[1,0].set_ylabel('Frequency')
    axes[1,0].set_title('Distribution of TOC')
    axes[1,0].grid(True, alpha=0.3)
    
    # Distribution of Water Content
    axes[1,1].hist(df['Water content'], bins=20, alpha=0.7, edgecolor='black', color='orange')
    axes[1,1].set_xlabel('Water Content (%)')
    axes[1,1].set_ylabel('Frequency')
    axes[1,1].set_title('Distribution of Water Content')
    axes[1,1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Main analysis
print("COMBINED DATASET ANALYSIS")
print("="*50)

# Combine all your datasets for overall analysis
# Assuming you have separate dataframes for each soil type
combined_data = []

# Add data from each soil type and layer
datasets = [
    (df_sand_city_0_30, 'Sand', '0-30'),
    (df_sand_city_30_100, 'Sand', '30-100'),
    (df_clay_city_0_30, 'Clay', '0-30'),
    (df_clay_city_30_100, 'Clay', '30-100'),
    (df_loam_city_0_30, 'Loam', '0-30'),
    (df_loam_city_30_100, 'Loam', '30-100'),
    (df_silt_city_0_30, 'Silt', '0-30'),
    (df_silt_city_30_100, 'Silt', '30-100')
]

for df, soil_type, layer in datasets:
    if len(df) > 0:  # Only add if dataframe has data
        temp_df = df[['Water content', 'TOC', 'Layer upper limit', 'Layer lower limit']].copy()
        temp_df['Soil type'] = soil_type
        temp_df['Layer'] = layer
        combined_data.append(temp_df)

# Create combined dataframe
df_combined = pd.concat(combined_data, ignore_index=True)

print(f"Total observations: {len(df_combined)}")
print(f"Soil types: {df_combined['Soil type'].unique()}")
print(f"Layers: {df_combined['Layer'].unique()}")

# Perform regression analysis
model_simple, model_multiple = perform_regression_analysis(df_combined)

# Create visualizations
plot_regression_results(df_combined, model_simple)

# Additional: Soil-type specific analysis
print("\n\n5. SOIL-TYPE SPECIFIC REGRESSION MODELS")
print("-" * 45)

for soil_type in df_combined['Soil type'].unique():
    soil_data = df_combined[df_combined['Soil type'] == soil_type]
    if len(soil_data) > 10:  # Only analyze if sufficient data
        X_soil = sm.add_constant(soil_data['Water content'])
        model_soil = sm.OLS(soil_data['TOC'], X_soil).fit()
        print(f"{soil_type}: R = {model_soil.rsquared:.3f}, Slope = {model_soil.params['Water content']:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')

def create_dummy_variables(df, reference_texture='Clay'):
    """
    Create dummy variables for soil texture and depth categories
    """
    df_analysis = df.copy()
    
    # Create depth variables
    # Continuous depth: midpoint of layer
    df_analysis['depth_continuous'] = (df_analysis['Layer upper limit'] + df_analysis['Layer lower limit']) / 2
    
    # Binary depth: topsoil (0-30cm) vs subsoil (30-100cm)
    df_analysis['depth_binary'] = np.where(df_analysis['Layer upper limit'] == 0, 0, 1)  # 0 = topsoil, 1 = subsoil
    
    # Create dummy variables for soil texture (one-hot encoding)
    texture_dummies = pd.get_dummies(df_analysis['Soil type'], prefix='texture')
    df_analysis = pd.concat([df_analysis, texture_dummies], axis=1)
    
    # Drop the reference category to avoid multicollinearity
    reference_col = f'texture_{reference_texture}'
    if reference_col in df_analysis.columns:
        df_analysis = df_analysis.drop(reference_col, axis=1)
        print(f"Using '{reference_texture}' as reference category (dropped {reference_col})")
    
    return df_analysis

def multiple_linear_regression_pandas(df):
    """
    Perform multiple linear regression using pandas and sklearn
    """
    print("="*80)
    print("MULTIPLE LINEAR REGRESSION ANALYSIS: TOC ~ Water Content + Depth + Texture")
    print("="*80)
    
    # Prepare data with dummy variables
    df_analysis = create_dummy_variables(df)
    
    # Model 1: Simple Linear Regression (TOC ~ Water Content)
    print("\n1. SIMPLE LINEAR REGRESSION: TOC ~ Water Content")
    print("-" * 50)
    
    X_simple = df_analysis[['Water content']]
    y = df_analysis['TOC']
    
    model_simple = LinearRegression()
    model_simple.fit(X_simple, y)
    
    y_pred_simple = model_simple.predict(X_simple)
    r2_simple = r2_score(y, y_pred_simple)
    
    print(f"R: {r2_simple:.4f}")
    print(f"Slope (): {model_simple.coef_[0]:.4f}")
    print(f"Intercept (): {model_simple.intercept_:.4f}")
    
    # Model 2: Multiple Regression with Continuous Depth
    print("\n2. MULTIPLE REGRESSION: TOC ~ Water Content + Depth (Continuous)")
    print("-" * 60)
    
    # Get all predictor columns (water content + depth + texture dummies)
    predictor_cols = ['Water content', 'depth_continuous']
    texture_cols = [col for col in df_analysis.columns if col.startswith('texture_')]
    predictor_cols.extend(texture_cols)
    
    X_multiple = df_analysis[predictor_cols]
    
    model_multiple = LinearRegression()
    model_multiple.fit(X_multiple, y)
    
    y_pred_multiple = model_multiple.predict(X_multiple)
    r2_multiple = r2_score(y, y_pred_multiple)
    
    print(f"R: {r2_multiple:.4f}")
    print("\nCoefficients:")
    for i, col in enumerate(predictor_cols):
        print(f"  {col}: {model_multiple.coef_[i]:.4f}")
    print(f"Intercept: {model_multiple.intercept_:.4f}")
    
    # Model 3: Multiple Regression with Binary Depth
    print("\n3. MULTIPLE REGRESSION: TOC ~ Water Content + Depth (Binary) + Texture")
    print("-" * 65)
    
    predictor_cols_binary = ['Water content', 'depth_binary']
    predictor_cols_binary.extend(texture_cols)
    
    X_multiple_binary = df_analysis[predictor_cols_binary]
    
    model_multiple_binary = LinearRegression()
    model_multiple_binary.fit(X_multiple_binary, y)
    
    y_pred_multiple_binary = model_multiple_binary.predict(X_multiple_binary)
    r2_multiple_binary = r2_score(y, y_pred_multiple_binary)
    
    print(f"R: {r2_multiple_binary:.4f}")
    print("\nCoefficients:")
    for i, col in enumerate(predictor_cols_binary):
        print(f"  {col}: {model_multiple_binary.coef_[i]:.4f}")
    print(f"Intercept: {model_multiple_binary.intercept_:.4f}")
    
    # Model Comparison
    print("\n4. MODEL COMPARISON")
    print("-" * 30)
    
    comparison_df = pd.DataFrame({
        'Model': ['Simple (Water only)', 'Multiple (Continuous Depth)', 'Multiple (Binary Depth)'],
        'R': [r2_simple, r2_multiple, r2_multiple_binary],
        'R Improvement': [0, r2_multiple - r2_simple, r2_multiple_binary - r2_simple]
    })
    
    print(comparison_df.to_string(index=False))
    
    # Interpretation
    print("\n5. INTERPRETATION OF RESULTS")
    print("-" * 35)
    
    # Water content coefficient interpretation
    water_coef_simple = model_simple.coef_[0]
    water_coef_multiple = model_multiple.coef_[0]
    water_coef_binary = model_multiple_binary.coef_[0]
    
    print(f"Water Content Coefficient:")
    print(f"  Simple model: {water_coef_simple:.4f} (TOC change per 1% water content)")
    print(f"  Multiple model (continuous): {water_coef_multiple:.4f}")
    print(f"  Multiple model (binary): {water_coef_binary:.4f}")
    
    # Depth coefficient interpretation
    depth_coef_continuous = model_multiple.coef_[1]
    depth_coef_binary = model_multiple_binary.coef_[1]
    
    print(f"\nDepth Coefficients:")
    print(f"  Continuous depth: {depth_coef_continuous:.4f} (TOC change per 1cm depth increase)")
    print(f"  Binary depth: {depth_coef_binary:.4f} (TOC difference: subsoil vs topsoil)")
    
    # Texture coefficients interpretation
    print(f"\nTexture Coefficients (relative to reference category):")
    for i, col in enumerate(texture_cols, 1):
        coef_continuous = model_multiple.coef_[1+i]  # +1 because water content and depth come first
        coef_binary = model_multiple_binary.coef_[1+i]  # +1 because water content and depth_binary come first
        texture_name = col.replace('texture_', '')
        print(f"  {texture_name}: {coef_continuous:.4f} (continuous), {coef_binary:.4f} (binary)")
    
    return {
        'df_analysis': df_analysis,
        'models': {
            'simple': model_simple,
            'multiple_continuous': model_multiple,
            'multiple_binary': model_multiple_binary
        },
        'predictor_cols': {
            'continuous': predictor_cols,
            'binary': predictor_cols_binary
        },
        'r2_scores': {
            'simple': r2_simple,
            'multiple_continuous': r2_multiple,
            'multiple_binary': r2_multiple_binary
        }
    }

def calculate_regression_statistics(df, model, predictor_cols, model_name):
    """
    Calculate additional regression statistics
    """
    X = df[predictor_cols]
    y = df['TOC']
    
    y_pred = model.predict(X)
    residuals = y - y_pred
    
    # Basic statistics
    n = len(y)
    p = len(predictor_cols)  # number of predictors
    r2 = r2_score(y, y_pred)
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
    rmse = np.sqrt(np.mean(residuals**2))
    
    print(f"\n{model_name} - Additional Statistics:")
    print(f"Number of observations: {n}")
    print(f"Number of predictors: {p}")
    print(f"Adjusted R: {adj_r2:.4f}")
    print(f"RMSE: {rmse:.4f}")
    
    return adj_r2, rmse

# Main analysis
print("COMBINED DATASET ANALYSIS WITH MULTIPLE REGRESSION")
print("="*60)

# Combine all datasets
combined_data = []

datasets = [
    (df_sand_city_0_30, 'Sand', '0-30'),
    (df_sand_city_30_100, 'Sand', '30-100'),
    (df_clay_city_0_30, 'Clay', '0-30'),
    (df_clay_city_30_100, 'Clay', '30-100'),
    (df_loam_city_0_30, 'Loam', '0-30'),
    (df_loam_city_30_100, 'Loam', '30-100'),
    (df_silt_city_0_30, 'Silt', '0-30'),
    (df_silt_city_30_100, 'Silt', '30-100')
]

for df, soil_type, layer in datasets:
    if len(df) > 0:
        temp_df = df[['Water content', 'TOC', 'Layer upper limit', 'Layer lower limit']].copy()
        temp_df['Soil type'] = soil_type
        temp_df['Layer'] = layer
        combined_data.append(temp_df)

df_combined = pd.concat(combined_data, ignore_index=True)

print(f"Dataset overview:")
print(f"Total observations: {len(df_combined)}")
print(f"Soil types: {df_combined['Soil type'].value_counts().to_dict()}")
print(f"Layers: {df_combined['Layer'].value_counts().to_dict()}")

# Perform multiple regression analysis
results = multiple_linear_regression_pandas(df_combined)

# Calculate additional statistics
print("\n" + "="*80)
print("DETAILED MODEL STATISTICS")
print("="*80)

calculate_regression_statistics(
    results['df_analysis'], 
    results['models']['simple'], 
    ['Water content'], 
    "Simple Linear Regression"
)

calculate_regression_statistics(
    results['df_analysis'], 
    results['models']['multiple_continuous'], 
    results['predictor_cols']['continuous'], 
    "Multiple Regression (Continuous Depth)"
)

calculate_regression_statistics(
    results['df_analysis'], 
    results['models']['multiple_binary'], 
    results['predictor_cols']['binary'], 
    "Multiple Regression (Binary Depth)"
)

# Create results summary dataframe
print("\n" + "="*80)
print("RESULTS SUMMARY DATAFRAME")
print("="*80)

summary_data = []
models_info = [
    ('Simple', results['models']['simple'], ['Water content']),
    ('Multiple_Continuous', results['models']['multiple_continuous'], results['predictor_cols']['continuous']),
    ('Multiple_Binary', results['models']['multiple_binary'], results['predictor_cols']['binary'])
]

for model_name, model, predictors in models_info:
    for i, predictor in enumerate(predictors):
        coef = model.coef_[i] if i < len(model.coef_) else np.nan
        summary_data.append({
            'Model': model_name,
            'Predictor': predictor,
            'Coefficient': coef,
            'R': results['r2_scores'].get(model_name.lower().replace('_', ''), np.nan)
        })
    # Add intercept
    summary_data.append({
        'Model': model_name,
        'Predictor': 'Intercept',
        'Coefficient': model.intercept_,
        'R': results['r2_scores'].get(model_name.lower().replace('_', ''), np.nan)
    })

summary_df = pd.DataFrame(summary_data)
print(summary_df.to_string(index=False))

# Save results to CSV
summary_df.to_csv('regression_results_summary.csv', index=False)
print("\nResults saved to 'regression_results_summary.csv'")
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Create depth categories
def categorize_depth(row):
    upper = row['Layer upper limit']
    lower = row['Layer lower limit']
    mid_depth = (upper + lower) / 2
    return 'Topsoil (0-30 cm)' if mid_depth <= 30 else 'Subsoil (30-100 cm)'

df_city['Depth_Category'] = df_city.apply(categorize_depth, axis=1)

# Separate data for topsoil and subsoil
topsoil_data = df_city[df_city['Depth_Category'] == 'Topsoil (0-30 cm)']
subsoil_data = df_city[df_city['Depth_Category'] == 'Subsoil (30-100 cm)']

print("=== Linear Regression by Depth Category ===")

# TOPSOIL Regression
if len(topsoil_data) > 1:
    X_top = topsoil_data[['Water content']]
    y_top = topsoil_data['TOC']
    
    model_top = LinearRegression()
    model_top.fit(X_top, y_top)
    
    r2_top = model_top.score(X_top, y_top)
    slope_top = model_top.coef_[0]
    intercept_top = model_top.intercept_
    
    print(f"\nTOPSOIL (0-30 cm):")
    print(f"  R: {r2_top:.4f}")
    print(f"  Equation: TOC = {slope_top:.4f} * Water + {intercept_top:.4f}")
    print(f"  Samples: {len(topsoil_data)}")

# SUBSOIL Regression
if len(subsoil_data) > 1:
    X_sub = subsoil_data[['Water content']]
    y_sub = subsoil_data['TOC']
    
    model_sub = LinearRegression()
    model_sub.fit(X_sub, y_sub)
    
    r2_sub = model_sub.score(X_sub, y_sub)
    slope_sub = model_sub.coef_[0]
    intercept_sub = model_sub.intercept_
    
    print(f"\nSUBSOIL (30-100 cm):")
    print(f"  R: {r2_sub:.4f}")
    print(f"  Equation: TOC = {slope_sub:.4f} * Water + {intercept_sub:.4f}")
    print(f"  Samples: {len(subsoil_data)}")

# Compare the relationships
if len(topsoil_data) > 1 and len(subsoil_data) > 1:
    print(f"\n=== Comparison ===")
    print(f"R difference: Topsoil ({r2_top:.4f}) vs Subsoil ({r2_sub:.4f})")
    print(f"Slope difference: Topsoil ({slope_top:.4f}) vs Subsoil ({slope_sub:.4f})")
    
    if r2_top > r2_sub:
        print(" Relationship is stronger in Topsoil")
    else:
        print(" Relationship is stronger in Subsoil")
```
