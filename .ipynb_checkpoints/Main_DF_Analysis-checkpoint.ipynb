````python
# drop null values in `Water content` and `TOC` 
df_clean = df_main.dropna(subset=["Water content", "TOC", "BD_bulk"])

# exclue any negative `Water content`
df_clean = df_clean[df_clean["Water content"] > 0]

print(len(df_clean))
````

## Add neccessary columns to using later, columns like `Soil type`, `Soil group`, `Layer group`
```
g/kg to %: Divide by 10 (since 10 g/kg = 1%)
% to g/kg: Multiply by 10
So the thresholds become:
For g/kg (your data):
Organic soils: TOC ≥ 300 g/kg (not 30!)
Mineral soils: TOC < 300 g/kg
```

````python
import numpy as np


# Smaller than `ORGANIC_THRESHOD` consider as mineral soil type
ORGANIC_THRESHOLD = 120.0

df_extra = df_clean.copy()


# Add `Soil type` column which is eigter `Mineral` or `Organic`
df_extra["Soil type"] = np.where(df_clean["TOC"] >= ORGANIC_THRESHOLD, "Organic", "Mineral")

# Add `Soil group` which is either `Sand` `Silt` `Clay` or `Loam`
textures = {
    "sand": ("Ss", "Su2", "Sl2", "Sl3", "St2", "Su3", "Su4", "Slu", "Sl4", "St3", "fS", "fSms", "mS", "mSgs", "gS", "mSfs"),
    "clay": ("Ts4", "Ts3", "Tu3", "Tu4", "Ts2", "Tl", "Tu2", "Tt", "Ts3"),
    "silt": ("Uu", "Us", "Ut2", "Ut3", "Uls", "Ut4"),
    "loam": ("Ls2", "Ls3",  "Ls4",  "Lt2",  "Lts", "Lu", "Lt3"),
}

def get_soil_group(texture):
    texture = texture.strip()
     # Check each category
    if texture in textures["sand"]:
        return 'Sand'
    elif texture in textures["clay"]:
        return 'Clay'
    elif texture in textures["silt"]:
        return 'Silt'
    elif texture in textures["loam"]:
        return 'Loam'
    else:
        return 'Other'

# Make the `Soil group` column
df_extra['Soil group'] = df_extra['Soil texture class'].apply(get_soil_group)

# Add `Layer group` Which is either `Topsoil` for layer `0-30` or `Subsoil` for layer in range `30-100`
df_extra['Layer group'] = np.where(
    (df_clean['Layer upper limit'] >= 0) & (df_clean['Layer lower limit'] <= 30), 
    'Topsoil',
    'Subsoil'
)


df_mineral = df_extra[df_extra["Soil type"] == "Mineral"]
df_organic = df_extra[df_extra["Soil type"] == "Organic"]


df_mineral["Soil group"].value_counts()
````

### Correlation for each layer

````python
from scipy.stats import pearsonr


def calculate_correlation_p(x, y):
    correlation, p_value = pearsonr(x, y)
    return round(correlation, 4), round(p_value, 4)


def determine_strength(correlation):
    # Determine strength and direction
    if abs(correlation) >= 0.7:
        strength = "strong"
    elif abs(correlation) >= 0.3:
        strength = "moderate"
    else:
        strength = "weak"
    return strength


def determine_direction(correlation):
    return "positive" if correlation > 0 else "negative"


def determine_significant(p_value):
    return p_value < 0.05


def create_correlation_table(data: dict[str, pd.DataFrame]):
    correlation_table =[] 
    for layer_name, df in data.items():
        x = df["Water content"]
        y = df["TOC"]
        corr, p = calculate_correlation_p(x, y)
        row = {
            "Layer group": layer_name,
            "correlation": corr,
            "p_value": p,
            "strength": determine_strength(corr),
            "direction": determine_direction(corr),
            "significant": determine_significant(p),
            "Mean TOC": round(df["TOC"].mean(), 1),
            "Mean Water c": round(df["Water content"].mean(), 1),
            "N samples": len(df),
        }
        correlation_table.append(row)
    df_correlation = pd.DataFrame(correlation_table)
    return df_correlation

# Mineral soil in different soil type.
df_mineral_group = {
    "0-30": df_mineral[df_mineral["Layer group"] == "Topsoil"], 
    "30-100": df_mineral[df_mineral["Layer group"] == "Subsoil"]
}
create_correlation_table(df_mineral_group)
````

### Correlation for each soil type

````python
soils = {
    "Sand": df_mineral[df_mineral["Soil group"] == "Sand"],
    "Clay": df_mineral[df_mineral["Soil group"] == "Clay"],
    "Silt": df_mineral[df_mineral["Soil group"] == "Silt"],
    "Loam": df_mineral[df_mineral["Soil group"] == "Loam"],
}

dfs_correlation = []
for soil_name, soil_df in soils.items():
    layer_group = {
        "0-30": soil_df[soil_df["Layer group"] == "Topsoil"],
        "30-100": soil_df[soil_df["Layer group"] == "Subsoil"],
    }
    res_df = create_correlation_table(layer_group)
    # Add soil name to the res df
    res_df.insert(0, "Soil group", soil_name)
    dfs_correlation.append(res_df)

result = pd.concat(dfs_correlation, ignore_index=True)
result
````

## Soil type

````python
import statsmodels.api as sm

# Define the independent variable (X) and dependent variable (y)
def print_ols(df, soil_name):
    X = df[['Water content', "BD_bulk"]]  # Independent variable
    y = df['TOC']            # Dependent variable
    
    # Add a constant to the model (intercept)
    X = sm.add_constant(X)
    
    # Fit the regression model
    model = sm.OLS(y, X).fit()

    print('-' * 20)
    print(soil_name)
    print('-' * 20)
    # Print the summary of the regression
    print(model.summary())


print_ols(soils["Sand"], "Sand")
print_ols(soils["Clay"], "Clay")
print_ols(soils["Silt"], "Silt")
print_ols(soils["Loam"], "Loam")
````
