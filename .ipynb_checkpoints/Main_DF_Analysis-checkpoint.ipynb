## Add neccessary columns to using later, columns like `Soil type`, `Soil group`, `Layer group`

```python
import numpy as np


# Smaller than `ORGANIC_THRESHOD` consider as mineral soil type
ORGANIC_THRESHOLD = 120.0

df_extra = df_clean.copy()


# Add `Soil type` column which is eigter `Mineral` or `Organic`
df_extra["Soil type"] = np.where(df_clean["TOC"] >= ORGANIC_THRESHOLD, "Organic", "Mineral")

textures = {
    "sand": ("Ss", "Su2", "Sl2", "Sl3", "St2", "Su3", "Su4", "fS", "fSms", "fSgs", "mS", "mSfs", "mSgs", "gS"),
    "loam": ("Slu", "Sl4", "St3", "Ls2", "Ls3", "Ls4", "Lt2", "Lts", "Ts4", "Ts3"),
    "silt": ("Uu", "Us", "Ut2", "Ut3", "Uls", "Ut4", "Lu"),
    "clay": ("Lt3", "Tu3", "Tu4", "Ts2", "Tl", "Tu2", "Tt"),
}

def get_soil_group(texture):
    texture = texture.strip()
     # Check each category
    if texture in textures["sand"]:
        return 'Sand'
    elif texture in textures["clay"]:
        return 'Clay'
    elif texture in textures["silt"]:
        return 'Silt'
    elif texture in textures["loam"]:
        return 'Loam'
    else:
        return 'Other'

# Make the `Soil group` column
df_extra['Soil group'] = df_extra['Soil texture class'].apply(get_soil_group)

# Add `Layer group` Which is either `Topsoil` for layer `0-30` or `Subsoil` for layer in range `30-100`
df_extra['Layer group'] = np.where(
    (df_clean['Layer upper limit'] >= 0) & (df_clean['Layer lower limit'] <= 30), 
    'Topsoil',
    'Subsoil'
)


df_mineral = df_extra[df_extra["Soil type"] == "Mineral"]
df_organic = df_extra[df_extra["Soil type"] == "Organic"]

df_mineral["Soil group"].value_counts()
# df_mineral[df_mineral["Soil group"] == "Other"]["Sand"]
```

### Correlation for each layer

```python
from scipy.stats import pearsonr


def calculate_correlation_p(x, y):
    correlation, p_value = pearsonr(x, y)
    return round(correlation, 4), round(p_value, 4)


def determine_strength(correlation):
    # Determine strength and direction
    if abs(correlation) >= 0.7:
        strength = "strong"
    elif abs(correlation) >= 0.3:
        strength = "moderate"
    else:
        strength = "weak"
    return strength


def determine_direction(correlation):
    return "positive" if correlation > 0 else "negative"


def determine_significant(p_value):
    return p_value < 0.05


def create_correlation_table(data: dict[str, pd.DataFrame]):
    correlation_table =[] 
    for layer_name, df in data.items():
        x = df["Water content"]
        y = df["TOC"]
        corr, p = calculate_correlation_p(x, y)
        row = {
            "Layer group": layer_name,
            "correlation": corr,
            "p_value": p,
            "strength": determine_strength(corr),
            "direction": determine_direction(corr),
            "significant": determine_significant(p),
            "Mean TOC": round(df["TOC"].mean(), 1),
            "Mean Water c": round(df["Water content"].mean(), 1),
            "N samples": len(df),
        }
        correlation_table.append(row)
    df_correlation = pd.DataFrame(correlation_table)
    return df_correlation

# Mineral soil in different soil type.
df_mineral_group = {
    "0-30": df_mineral[df_mineral["Layer group"] == "Topsoil"], 
    "30-100": df_mineral[df_mineral["Layer group"] == "Subsoil"]
}
create_correlation_table(df_mineral_group)



```

```python
soils = {
    "Sand": df_mineral[df_mineral["Soil group"] == "Sand"],
    "Clay": df_mineral[df_mineral["Soil group"] == "Clay"],
    "Silt": df_mineral[df_mineral["Soil group"] == "Silt"],
    "Loam": df_mineral[df_mineral["Soil group"] == "Loam"],
}

dfs_correlation = []
for soil_name, soil_df in soils.items():
    layer_group = {
        "0-30": soil_df[soil_df["Layer group"] == "Topsoil"],
        "30-100": soil_df[soil_df["Layer group"] == "Subsoil"],
    }
    res_df = create_correlation_table(layer_group)
    # Add soil name to the res df
    res_df.insert(0, "Soil group", soil_name)
    dfs_correlation.append(res_df)

result = pd.concat(dfs_correlation, ignore_index=True)
result
```

### Correlation for soil texture class

```python
# pd.set_option('display.max_rows', None)

def create_correlation_table_texture(data: dict[str, pd.DataFrame]):
    correlation_table =[] 
    for layer_name, df in data.items():
        x = df["Water content"]
        y = df["TOC"]
        corr, p = calculate_correlation_p(x, y)
        row = {
            "Layer group": layer_name,
            "correlation": corr,
            "p_value": p,
            "strength": determine_strength(corr),
            "direction": determine_direction(corr),
            "significant": determine_significant(p),
            "Mean TOC": round(df["TOC"].mean(), 1),
            "Mean Water c": round(df["Water content"].mean(), 1),
            "N samples": len(df),
        }
        correlation_table.append(row)
    df_correlation = pd.DataFrame(correlation_table)
    return df_correlation
    
def texture_table():
    # df_mineral = df_mineral["Soil texture class"].dropna()
    textures = df_mineral['Soil texture class'].unique()
    texture_dfs = []
    for texture in textures:
        texture_df = df_mineral[df_mineral['Soil texture class'] == texture]     
        if len(texture_df) > 3:
            layer_group = {
                "0-30": texture_df[texture_df["Layer group"] == "Topsoil"],
                "30-100": texture_df[texture_df["Layer group"] == "Subsoil"],
            }
            corr_texture = create_correlation_table_texture(layer_group)
            corr_texture["Texture class"] = texture 
        
            texture_dfs.append(corr_texture)
    return texture_dfs


dfs_textures = texture_table()    

result_t = pd.concat(dfs_textures, ignore_index=True)
result_t
```

## Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_layer_scatterplots(df, layer_ranges, soil_name="", x_col='Water content', y_col='TOC', figsize=(10, 6), colors=None):
    # Create layer DataFrames
    layer_dfs = []
    for upper, lower in layer_ranges:
        df_layer = df[(df["Layer upper limit"] == upper) & (df["Layer lower limit"] <= lower)]
        layer_dfs.append((f"{upper}_{lower}", df_layer))
    
    # Set default colors if not provided
    if colors is None:
        colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown']
    
    # Find global axis limits
    x_min, x_max = float('inf'), float('-inf')
    y_min, y_max = float('inf'), float('-inf')
    
    for name, df_layer in layer_dfs:
        if len(df_layer) > 0:
            x_min = min(x_min, df_layer[x_col].min())
            x_max = max(x_max, df_layer[x_col].max())
            y_min = min(y_min, df_layer[y_col].min())
            y_max = max(y_max, df_layer[y_col].max())
    
    # Add padding
    x_padding = (x_max - x_min) * 0.1 if x_max > x_min else 0.1
    y_padding = (y_max - y_min) * 0.1 if y_max > y_min else 0.1
    
    x_min -= x_padding
    x_max += x_padding
    y_min -= y_padding
    y_max += y_padding
    
    # Create subplots
    n_plots = len(layer_dfs)
    n_cols = min(3, n_plots)
    n_rows = (n_plots + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]
    
    # Plot each layer
    for i, (name, df_layer) in enumerate(layer_dfs):
        ax = axes[i]
        
        if len(df_layer) > 0:
            # Scatter plot
            ax.scatter(df_layer[x_col], df_layer[y_col], color=colors[i], alpha=0.7)
            
            # Linear regression
            if len(df_layer) > 1:
                slope, intercept = np.polyfit(df_layer[x_col], df_layer[y_col], 1)
                regression_line = slope * df_layer[x_col] + intercept
                ax.plot(df_layer[x_col], regression_line, color='red', 
                       label=f'y = {slope:.2f}x + {intercept:.2f}')
            
            ax.set_title(f'{soil_name} in Layer {name.replace("_", "-")} cm')
            ax.set_xlabel(x_col)
            ax.set_ylabel(y_col)
            ax.grid(True, alpha=0.3)
            ax.legend()
            
            # Set consistent axis limits
            ax.set_xlim(x_min, x_max)
            ax.set_ylim(y_min, y_max)
        else:
            ax.text(0.5, 0.5, 'No data available', ha='center', va='center', 
                   transform=ax.transAxes)
            ax.set_title(f'{soil_name} in Layer {name.replace("_", "-")} cm')
            ax.set_xlim(x_min, x_max)
            ax.set_ylim(y_min, y_max)
    
    # Remove empty subplots
    for j in range(len(layer_dfs), len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    return fig, axes

```

### Plot for each soil in topsoil and subsoil layers.

```python
# Usage example:
layer_ranges = [(0, 30), (30, 100)]
for soil_name in ["Sand", "Clay", "Silt", "Loam"]:
    fig, axes = plot_layer_scatterplots(soils[soil_name], layer_ranges, soil_name=soil_name)
    plt.show()

```
