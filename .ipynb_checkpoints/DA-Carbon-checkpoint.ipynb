```python
import pandas as pd


pd.set_option('display.max_columns', None)
df = pd.read_csv("LABORATORY_DATA.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# Data Cleaning

```python
df_da = df[["County", "TOC", "Water content"]]

df_da.dropna(inplace=True)

df_da.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Your existing code
plt.figure(figsize=(10, 6))
plt.scatter(df_da['TOC'], df_da['Water content'], color='blue')

# Calculate linear regression
slope, intercept = np.polyfit(df_da['TOC'], df_da['Water content'], 1)

# Create regression line
regression_line = slope * df_da['TOC'] + intercept

# Plot regression line
plt.plot(df_da['TOC'], regression_line, color='red', label=f'Linear Regression: y = {slope:.2f}x + {intercept:.2f}')

plt.title('Scatter Plot of TOC vs Water Content with Linear Regression')
plt.xlabel('Total Organic Carbon (TOC)')
plt.ylabel('Water Content')
plt.grid()
plt.legend()
plt.show()
```

```python
# correlation
# Step 3: Calculate Correlation
correlation = df['TOC'].corr(df['Water content'])
print(f'Correlation between TOC and Water Content: {correlation:.2f}')
```

# Result of above code

Implication: **As the TOC increases, the water content also tends to increase.** This suggests that higher levels of organic carbon in the soil or substrate are associated with higher water retention or content.

Soil Health: In agricultural or environmental contexts, **this relationship can imply that soils with higher organic carbon content are better at retaining moisture, which is beneficial for plant growth and ecosystem health.**
